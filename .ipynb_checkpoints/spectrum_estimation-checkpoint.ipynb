```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lambd = np.linspace(300,1000,10000)*10**(-9)
v = 3*10**8/lambd[::-1]
T = np.max(v)-np.min(v)
N = 2
fn = np.array(range(N))+1
F = fn[0]
for n in range(1,N):
    F = F + fn[n]*np.cos(v*n*np.pi/T)
plt.figure()
plt.plot(v,F)
```

```python
#If this accurate could exactly compute integrals
x = 1.065*np.exp(-0.5*((lambd*10**9-595.8)/33.33)**2)+0.366*np.exp(-0.5*((lambd*10**9-446.8)/19.44)**2)
plt.figure()
plt.plot(lambd*10**9,lambd*x)
```

```python
K = 20
theta = np.pi*np.array(range(K))/(6*K)
ctheta = np.cos(theta)

#measurements
X=np.zeros(K)
for k in range(K):
    X[k] = np.sum(np.interp(v*ctheta[k], v, x)*F)
X = X + 10*np.random.randn(len(X))

S = np.zeros([K,N])
for k in range(K):
    for n in range(N):
        S[k,n] = np.sum(np.interp(v*ctheta[k], v, x)*np.cos(n*v*np.pi/T))

print np.linalg.cond(S)
#print S
#print X
#print S.dot(np.array([1,2]))

print np.linalg.inv(S.transpose().dot(S)).dot(S.transpose().dot(X))
```
